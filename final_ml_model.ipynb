```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
```

```python
df.head()
```

```python
le = preprocessing.LabelEncoder()
x1 = df["SOIL"]
y1 = df["STATE"]
x = le.fit_transform(x1)
y = le.fit_transform(y1)
df1 = df.merge(pd.DataFrame({'SOIL_C':x,'STATE_C':y}),right_index=True,left_index = True)
df1.head()
```

```python
print("For State : {} and Region : {}".format(state,region))
ordinary=df1.query('STATE == @state and Region_C == @region')
reg_soil=ordinary[['STATE_C','SOIL_C']].values #input for classifier
soil_list = list(ordinary['SOIL_C'].values)
#output for state and region
soil_type=ordinary[['SOIL']].values
print("Soil Types found are : \n",soil_type)
```

```python
basic=pd.read_excel('/content/season_crop.xlsx')
```

```python
basic.head()
```

```python
le2 = preprocessing.LabelEncoder()
x2 = basic["SOIL"]
y2 = basic["STATE"]
z2 = basic["SEASON"]
a = le2.fit_transform(x2)
b = le2.fit_transform(y2)
c= le2.fit_transform(z2)
basic1 = basic.merge(pd.DataFrame({'SOIL_C':a,'STATE_C':b,'SEASON_C':c}),right_index=True,left_index = True)
basic1.head()
```

```python
X = basic1[['STATE_C','SOIL_C','SEASON_C']]
Y = basic1[['Rice','Cotton','Sugarcane','Wheat','Millets','Cardamom','Ginger','Coconut','Orange','Soyabean','Maize']]
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X,Y)
```
